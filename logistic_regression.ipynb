Логістична регресія - для бінарної класифікації

Визначається Сігмоід функцією

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sigmoid f(x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# похідна
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
# генерація значень від -6 ~ 6
x = np.linspace(-6, 6, 100)
x
```

```python
# обчислюєм значення сігмоід ф-ції та похідної кожного x
y_sigmoid = sigmoid(x)
y_derivative = sigmoid_derivative(x)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1) # 2row, 1col
plt.plot(x, y_sigmoid)
plt.title("Sigmoid f(x)")
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x, y_derivative, color='r')
plt.title("Sigmoid f(x) derivative")
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
